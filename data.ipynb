```python
def uniquie_percent_count(data, col):
    col_count = data[col].value_counts(normalize = True)
    col_count = pd.DataFrame(col_count)
    col_count = col_count.reset_index()
    col_count = col_count.rename(columns={col:col, 'proportion':'percentage_of_car'})
    col_count['percentage_of_car'] = col_count['percentage_of_car']*100
    col_count['percentage_of_car']= col_count["percentage_of_car"].round(2)
    return col_count
    
```

```python
fuelType_count = uniquie_percent_count(df, 'fuelType')
display(fuelType_count)
```

```python
transmission_count = uniquie_percent_count(df, 'transmission')
display(transimission_count)
```

```python
model_count = uniquie_percent_count(df, 'model')
display(model_count)
```

```python
year_count = uniquie_percent_count(df, 'year')
year_count = year_count.sort_values(by='year')
display(year_count)
```

```python
def barplot(data, x_col, y_col, color, yticks, rotation):
    sns.barplot(x= x_col, y = y_col, data = data, color= color)
    plt.title('percentage of car presenet for each '+ 'x_col')
    plt.yticks(yticks)
    plt.xticks(rotation = rotation)
    
```

```python
sns.set_context('paper')  #paper, talk, poster

plt.figure(figsize=(18,20))

plt.subplot(2,2,1)
barplot(model_count, 'model', 'percentage_of_car', 'orange', np.arange(0,41,20), 90)

plt.subplot(2,2,2)
barplot(year_count, 'year', 'percentage_of_car', 'black', np.arange(0,41,20), 90)

plt.subplot(2,2,3)
barplot(transmission_count, 'transmission', 'percentage_of_car', 'orange', np.arange(0,71,20), 0)

plt.subplot(2,2,4)
barplot(fuelType_count, 'fuelType', 'percentage_of_car', 'blue', np.arange(0,71,20), 0)


plt.show()
#plt.tight_layout()
```

Creating trend line

```python
plt.figure(figsize=(20,12))
plt.subplot(3,1,1)
sns.regplot(data=df, x='mileage', y='price', color='r')

plt.subplot(3,1,2)
sns.regplot(data=df, x='mileage', y='price', line_kws={'color':'r'})

plt.subplot(3,1,3)
sns.regplot(data=df, x='mileage', y='price', scatter_kws={'color':'r', 'edgecolor':'white'})

plt.tight_layout()
plt.show()
```
