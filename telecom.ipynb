```python
df = pd.read_csv("telecom_data.csv")
```

**Histogram plot**

```python
plt.figure(figsize=(18,12))

plt.subplot(3,3,1)
sns.histplot(data= df, x= 'Total_calls')
plt.xticks(np.arange(191,417,20))

plt.subplot(3,3,2)
sns.histplot(data= df, x= 'Total_calls', binwidth = 50)
plt.xticks(np.arange(191,417,20))

plt.subplot(3,3,3)
sns.histplot(data= df, x= 'Total_calls', binwidth = 20, color="green")
plt.xticks(np.arange(191,417,20))

plt.subplot(3,3,4)
sns.histplot(data= df, x= 'Total_calls', binwidth = 20, color="orange")
plt.xticks(np.arange(191,417,20))

plt.subplot(3,3,5)
sns.histplot(data= df, x= 'Total_calls', binwidth = 20, color="orange")
plt.xticks(np.arange(191,417,20))


plt.subplot(3,3,6)
sns.histplot(data= df, x= 'Total_calls', binwidth = 20, color="green", fill=False)
plt.xticks(np.arange(191,417,20))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18,12))

plt.subplot(3,3,1)
sns.histplot(data= df, x= 'Total_calls', binwidth = 1)
plt.xticks(np.arange(191,417,20))

plt.subplot(3,3,2)
sns.histplot(data= df, x= 'Total_calls', binwidth = 1, cumulative=True)
plt.xticks(np.arange(191,417,20))

plt.subplot(3,3,3)
sns.histplot(data= df, x= 'Total_calls', binwidth = 1, cumulative=True, stat='probability')
plt.xticks(np.arange(191,417,20))
plt.grid()

plt.subplot(3,3,4)
sns.histplot(data= df, x= 'Total_calls', binwidth = 1, kde=True)

plt.subplot(3,3,5)
sns.histplot(data= df, x= 'Total_calls', binwidth = 1, kde=True, hue='Churn')

plt.subplot(3,3,6)
sns.histplot(data= df, x= 'Total_calls', binwidth = 1, kde=True, element='poly')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18,12))

plt.subplot(1,3,1)
sns.boxplot(x='Churn', y="Total_charges", data=df, showfliers=False)

plt.subplot(1,3,2)
sns.violinplot(x='Churn', y="Total_charges", data=df)

plt.subplot(1,3,3)
sns.violinplot(x='Churn', y="Total_mins", data=df)

plt.show()
```
